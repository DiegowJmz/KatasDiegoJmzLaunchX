1er Ejercicio

```python
# Creamos nuestras variables necesarias

pPlaneta = 149597870
sPlaneta = 778547200

# Ahora buscamos la diferencia entre estod dos planetas y la guardamos en una variable

print ("La tierra esta a " + str(pPlaneta) + " y Jupiter a " + str(sPlaneta) + " de distancia")

dPlanetaria = sPlaneta - pPlaneta

# Procedemos a imprimir 

print ("La distancia entre ambos planetas es: " + str(dPlanetaria)) 

# Usando el tip que nos dio en la kata, multiplicamos para pasar a millas. 

mPlanetaria = dPlanetaria * 0.621 

print ("La distancia en millas entre ambos planetas es: " + str(mPlanetaria))

```
