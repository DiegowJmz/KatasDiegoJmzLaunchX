Ejercicio 1

```python
# Creamos la lista con los 8 planetas
i = 1
planets = ["Mercurio", "Venus", "Tierra", "Marte", "Jupiter", "Saturno", "Urano", "Neptuno"]

# Usamos un ciclo for para poder mostrar todos los planetas, junto con la posicion en la que se encuntra
for n in planets:
    print ("El planeta " + n + " esta en la posicion " + str(i))
    i = i+1

# Guardamos la palabra pluton en una variable, para despues agregarla a la lista planets

planetAdd = 'pluton' 

# Usando append agregamos la variable planetAdd a la lista planets

planets.append(planetAdd)

# Ahora mostramosa la lista de nuevo con pluton agregado

j = 1
print ( "- - - - - - - - - - - - - - - - - - - - - - - - ")
for n in planets:
    print ("El planeta " + n + " esta en la posicion " + str(j))

    j = j+1
```

Ejercico 2

```python
planets = ["Mercurio", "Venus", "Tierra", "Marte", "Jupiter", "Saturno", "Urano", "Neptuno"]

bPlaneta = input("Introduce el planeta que quieres buscar ")

# Ahora usamos index para buscar en la lista. 

ePlaneta = planets.index(bPlaneta)

# Ahora, mostraremos el primer planeta y el planeta que indico el usuario

print("Aqui esta el planetas mas cerca que " + bPlaneta)
print(planets[0:ePlaneta])

```
